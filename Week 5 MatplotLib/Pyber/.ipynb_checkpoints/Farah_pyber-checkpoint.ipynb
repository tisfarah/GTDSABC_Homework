### Import, Merge, and Explore Data

```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Read the City and Ride Data
city_data = pd.read_csv("data/city_data.csv")
ride_data = pd.read_csv("data/ride_data.csv")

# Combine the data into a single dataset
city_ride = pd.merge(ride_data, city_data, on="city", how="left")
# Display the data table for preview
city_ride.head()

# don't care about date column, so we'll drop that column 
```

### Create a DataFrame 

```python
city_group = city_ride.groupby("city").agg({"fare":"mean", "ride_id":"count",
                                                    "driver_count" : "mean"})

city_group.sort_values("ride_id")

update_cityride = pd.merge(city_group, city_data, how="left", on=["city", "city"])
rides = update_cityride[["city", "fare", "ride_id","driver_count_x", "driver_count_y", "type"]]

rides["colors"] = ["Gold" if x == "Rural" else "SkyBlue" if x == "Suburban"  else "Coral" for x in rides["type"]]
rides.head()

```

## Task 1: Bubble Plot of Ride Sharing Data

Build a Bubble Plot that showcases the relationship between four key variables:
* Average Fare ($) Per City
* Total Number of Rides Per City
* Total Number of Drivers Per City
* City Type (Urban, Suburban, Rural)


```python
# Obtain the x and y coordinates for each of the three city types

# X coordinates are the total number of rides by city by driver count 
x = rides["ride_id"]

# Y coordinates are the average fares 
y = rides["fare"]

# the size of the bubbles (this is what differentiates this graph from scatter) is driver count
s = 10 * rides["driver_count_y"]

colors = rides["colors"]
c_type = [x for x in rides["type"].unique()] 
# Build the scatter plots for each city types
plt.subplots()

# fig, ax = plt.subplots()

plt.scatter(x, y, s,c=colors, edgecolors="black", alpha =.75, linewidths = 1, label = c_type)


# Incorporate the other graph properties
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.title("Pyber Ride Sharing Data (2018)")
plt.grid(True)

# Create a legend
plt.legend()
# ax.legend(loc='upper right', title =  "City Types" )

# https://matplotlib.org/examples/lines_bars_and_markers/scatter_with_legend.html

# Incorporate a text label regarding circle size

plt.text(1.02,.7, 'Note: \nCircle size correlates with driver count per city.', ha='left', va='center', transform=ax.transAxes)


# Save Figure
plt.show()

# plt.savefig("PyberBubble_Farah.png")
```

## Task 2: Total Fares by City Type

```python
# Calculate Fare Percents
fare_type = city_ride.groupby("type").agg({"fare":"sum"})


total_fare = fare_type.sum()
rural_fare = fare_type["fare"]["Rural"]
urban_fare = fare_type["fare"]["Urban"]
sub_fare = fare_type["fare"]["Suburban"]
fare_type

labels = ["Urban",  "Rural","Suburban",]
sizes = [urban_fare, rural_fare, sub_fare ]
colors = ["Salmon", "Gold", "SkyBlue"]
explode = (0.1, 0, 0.0)

# Build Pie Chart
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
      autopct="%1.1f%%",  shadow=True, startangle=300 )

plt.title("% of Total Fares by City Type")
# plt.axis("equal")
#
# Save Figure
plt.savefig("CityFarePie_Farah.png")
```

## Task 3: Total Rides by City Type

```python
# Calculate Rides Percents
rides_type = city_ride.groupby("type").agg({"ride_id":"count"})


rural_rides= rides_type["ride_id"]["Rural"]
urban_rides = rides_type["ride_id"]["Urban"]
sub_rides = rides_type["ride_id"]["Suburban"]


labels = ["Urban",  "Rural","Suburban",]
sizes = [urban_rides, rural_rides, sub_rides]
colors = ["Salmon", "Gold", "SkyBlue"]
explode = (0.1, 0, 0.0)

# Build Pie Chart
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
      autopct="%1.1f%%",  shadow=True, startangle=270 )

plt.title("% of Total Rides by City Type")
# plt.axis("equal")
#
# Save Figure
plt.savefig("CityRidePie_Farah.png")
```

## Task 4: Total Drivers by City Type

```python
# Calculate Driver Percents
# Use City Data since that dataset does to have duplicates
drivers = city_data.groupby("type").sum()


rural_dr= drivers["driver_count"]["Rural"]
urban_dr = drivers["driver_count"]["Urban"]
sub_dr = drivers["driver_count"]["Suburban"]


labels = ["Urban",  "Rural","Suburban",]
sizes = [urban_dr, rural_dr, sub_dr]
colors = ["Salmon", "Gold", "SkyBlue"]
explode = (0.2, 0, 0.0)

# Build Pie Chart
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
      autopct="%1.1f%%",  shadow=True, startangle=220 )

plt.title("% of Total Drivers by City Type")
# plt.axis("equal")
#
# Save Figure
plt.savefig("CityDriverPie_Farah.png")
```
